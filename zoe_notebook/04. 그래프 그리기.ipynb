### 04-1. 데이터 시각화가 필요한 이유 ###

**앤스콤 데이터 집합 불러온 후 그래프 그리기**

```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

# 데이터 값이 I인 첫번째 데이터 그룹만 추출
dataset_1 = anscombe[anscombe['dataset'] == 'I']
```

```python
plt.plot(dataset_1['x'], dataset_1['y'])
```

```python
# 세번째 인자에 'o' 값을 주면 그래프가 점으로 그려진다. 
plt.plot(dataset_1['x'], dataset_1['y'], 'o')
```

**matplotlib 라이브러리로 그래프 그리기**  
1. 전체 그래프가 위치할 기본 틀 만들기
2. 그래프를 그려 넣을 그래프 격자 만들기
3. 격자에 그래프 하나씩 추가하기. 격자에 그래프가 추가되는 순서는 왼->오
4. 만약 격자의 첫번째 행이 꽉 차면 두번째 행에 그래프 그리기

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
# 그래프 격자가 위치할 기본 틀 만들기
fig = plt.figure()
```

```python
# add_subplt 메서드로 그래프 격자 그리기
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
```

```python
# plot 메서드에 데이터를 전달하여 그래프 그리기
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')
```

```python
# 그래프 격자에 제목 추가
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")
```

```python
# 기본 틀(fig)에도 제목 추가
fig.suptitle("Anscombe Data")
```

```python
# 그래프의 이름과 숫자가 겹쳐보여서 레이아웃 조정
fig.tight_layout()
```

### 04-2. matplotlib 라이브러리 자유자재로 사용하기 ###

```python
# tips 데이터집합 불러오기
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

**일변량 그래프** : 변수를 하나만 사용해서 그린 그래프

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)

# bins로 x축의 간격 조정 가능
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

**산점도 그래프** : 변수 2개를 사용해서 만드는 그래프로, 변수 2개를 사용하기 때문에 통계용어로 '이변량 그래프'라고 부른다.

```python
# total_bill 열에 따른 tip 열의 분포를 나타낸 산점도 그래프
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

**박스그래프** : 이산형 변수와 연속형 변수를 함께 사용하는 그래프  
**이산형 변수**란 Female, Male과 같이 명확하게 구분되는 값을 의미하고, **연속형 변수**란 Tip과 같이 명확하게 셀 수 없는 범위의 값을 의미한다.

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
             labels=['Female', 'Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

**다변량 그래프** : 3개 이상의 변수를 사용한 그래프  
적재적소에 맞는 그래프 요소를 추가하여 그래프를 표현해야 한다.

```python
# 산점도 그래프에 새 변수 '성별' 추가하기
# 'Female', 'Male' 문자열을 각각 정수로 치환
# Female : 0, Male : 1

def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,     # 점의 크기 : 테이블당 인원 수
    c=tips['sex_color'],   # 점의 색상 : 성별
    alpha=0.5              # 점의 투명도
)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### 04-3. seaborn 라이브러리 자유자재로 사용하기 ###

**단변량 그래프 그리기 - 히스토그램**

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots()
# distplot 메서드를 사용하면 히스토그램과 밀집도 그래프를 같이 그릴 수 있다.
ax = sns.distplot(tips['total_bill'])
# 만약 밀집도 그래프를 제외하고 싶다면 kde 인자값을 False로 설정하면 된다.
# ax = sns.distplot(tips['total_bill'], kde=False)
# 밀집도 그래프만 나타내려면 hist 인자를 False로 지정하면 된다.
# ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Histogram With Density Plot')
```

```python
# 데이터 시각화를 더 효율적으로 하기위해 rug 인자를 추가할 수도 있다.  
# 양탄자 그래프는 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래프이다.
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
## count 그래프는 이산값을 나타낸 그래프이다.
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

**seaborn 라이브러리로 산점도 그래프 그리기**

```python
# regplot 메서드를 사용하면 산점도 그래프와 회귀선을 함께 그릴 수 있다.
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
# 만약 회귀선을 제거하려면 fit_reg 인자를 False로 지정하면 된다.
# ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 산점도 그래프와 히스토그램을 한번에 그려주는 jointplot 메서드
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 산점도 그래프는 점이 겹쳐보일 경우 구분하기 어렵다는 단점이 있다.
# 산점도 그래프의 데이터를 쉽게 구분하기 위해서 육각 그래프를 사용하면 된다.
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind="hex")
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 이차원 밀집도 그리기
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                data2=tips['tip'],
                shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 바 그래프 그리기 : 지정한 변수의 평균을 계산하여 그릴 수 있다.
# 시간에 따라 지불한 비용의 평균
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
# 박스 그래프 : 최소값, 1분위수, 중간값, 3분위수, 최대값, 이상치 등 다양한 통계량으 한번에 표현하는 그래프
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

```python
# 바이올린 그래프 : 박스그래프는 다양한 통계수치를 확인하기 위해 자주 사용하는 그래프지만 데이터 분산이 모호하게 표현된다.
# 이런 경우에 박스 그래프에 커널 밀도를 추정한 바이올린 그래프를 사용한다.
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```
